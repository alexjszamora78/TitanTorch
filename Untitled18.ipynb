```python
from sklearn.linear_model import LinearRegression

# Creamos un objeto de regresión lineal
model = LinearRegression()

# Dividimos los datos en conjuntos de entrenamiento y prueba
X_train = [[5,3,1,1,0,1,8,5,1,3,9,4,7,6,2,0,1,2,5,90,91,92,93,94,95,96,7,98,99,1,3,8,5,28,84,20,75,33]]
y_train = [98]
X_test = [[5,3,1,1,0,1,8,5,1,3,9,4,7,6,2,0,1,2,5,90,91,92,93,94,95,96,7,98,99,1,3,8,5,28,84,20,75,33]]
y_test = [51]

# Ajustamos el modelo a los datos de entrenamiento
model.fit(X_train, y_train)

# Predecimos los resultados para los datos de prueba
y_pred = model.predict(X_test)

# Imprimimos las predicciones
print(y_pred)
```
